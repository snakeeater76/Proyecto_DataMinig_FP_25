```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Correctly encode 'SAC' column to sequential class labels
# Assume dfimport is available from previous cells
le = LabelEncoder()
y_encoded = le.fit_transform(dfimport['SAC'].values)
num_classes = len(le.classes_)

# Re-define X and fill NaN values to ensure consistency
X = dfimport[['ANYO', 'MES', 'PAIS', 'ADUANA', 'VIA', 'PESO', 'VALOR']].copy()
X = X.fillna(0)

# Re-run train_test_split with the corrected y_encoded
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Re-define the model with the correct number of output classes
model = Sequential()
model.add(Dense(8192, input_dim=7, activation='relu'))
model.add(Dense(8192, activation='relu'))
model.add(Dense(num_classes, activation='softmax')) # Corrected num_classes

# Re-compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Finally, fit the model
model.fit(X_train, y_train, epochs=50, batch_size=64, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'loss:{loss}, Accuracy:{accuracy}')
```

```python
hecho = np.array([[1,44,3,44,1,3]])
res = model.predict(hecho)
print(res)
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
data3 = pd.read_csv('/content/drive/MyDrive/DataMinig/Estadistica de agresiones 24.csv')
```

```python
data3.head()
```

```python
data3['V_ACTIVO'] = data3['VIC_DEDICA'].apply(lambda x: 1 if x < 4 else 0)
```

```python
data3 = data3.sample(frac=1, random_state=42,).reset_index(drop=True)
```

##Variables Independientes y Dependientes

```python
x = data3[['AGR_ESCOLARIDAD','AGR_SEXO', 'AGR_TRABAJA', 'VIC_SEXO']]
y = data3['V_ACTIVO']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

RED NEURONAL

```python
model = Sequential()
model.add(Dense(6, input_dim=4, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=50, batch_size=64, validation_data=(x_test, y_test))
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'loss:{loss}, Accuracy:{accuracy}')
```

PREDICCION: Se busca predecir, si la victima se encuentra productiva (Que haceres casa, Vive de rentas o estudia), con base en las caracteristicas del agresor (Escolaridad, Sexo y si Trabaja)

```python
vic_prod = np.array([[21, 1, 2, 1]])
resul = model.predict(vic_prod)
print(resul)
```
